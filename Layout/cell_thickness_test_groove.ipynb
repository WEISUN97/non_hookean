```python
import sys
sys.path.append("/Users/bubble/Desktop/Project/Buckling/non_hookean/Layout")
import math
import gdsfactory as gf
from gdsfactory.generic_tech import get_generic_pdk
PDK = get_generic_pdk()
PDK.activate()
cell_temp = gf.Component()
```

```python
'''
layer 1: thick ald layer
layer 2: thin ald layer
layer 3: KOH layer
layer 4: KOH (backside asymmetric) layer
'''
```

```python
# get the horizontal distance caused by KOH etching.
def KOHSize(height):
    size = height / math.tan(math.radians(54.74))
    return round(size, 2)
# print(KOHSize(525)*2)

block = gf.Component()
# backside etching pattern
gap = 2000
length_etch = 14000 + KOHSize(525)*2
width_etch = 500 + KOHSize(525)*2
rec_etch = gf.components.rectangle(size=(length_etch, width_etch), layer=(3, 0))
for i in range(3):
    (block << rec_etch).move((8000-length_etch/2, -1200 + i*width_etch + i*gap))

    
# block.show()
```

```python
# boolean operation
# do backside etching - frame
cell_temp << block
cell_temp2 = gf.Component()
(cell_temp2 << cell_temp).move((-500, (10830-5000)/2))
# frame
frame1 = gf.components.rectangle(size=(15000, 10830), layer=(20, 0))
frame2 = gf.components.rectangle(size=(20000, 15830), layer=(21, 0))
frame2_ref = cell_temp2 << frame2
frame2_ref.move((-2500, -2500))
cell_temp2 << frame1

cell_thickness_test_groove = gf.Component()
cell_ref = cell_thickness_test_groove << cell_temp2
cell_ref.move((2500, -7500))
# cell_thickness_test_groove.show()
# cell_thickness_test_groove.write_gds("mesh.gds")
# cell_thickness_test_groove.plot()
```
