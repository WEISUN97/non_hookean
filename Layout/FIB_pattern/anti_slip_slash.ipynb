```python
%reset -f
# test piece for non-hookean metamaterial
import gdsfactory as gf
import import_ipynb
from datetime import datetime
import sys
import math
sys.path.append("/Users/bubble/Desktop/Project/Buckling/non_hookean/Layout")
from gdsfactory.generic_tech import get_generic_pdk
PDK = get_generic_pdk()
PDK.activate()
```

```python
# frame
factor = 4
frame = gf.components.rectangle(size=(10, 110), layer=(1, 0))
rec_1 = gf.components.rectangle(size=(1/factor, 5/factor), layer=(2, 0))
rec_2 = gf.components.rectangle(size=(5/factor, 1/factor), layer=(2, 0))
pattern_unit = gf.Component()
(pattern_unit << rec_1).move((-3/factor, -5/factor))
(pattern_unit << rec_1).move((2.5/factor, 0))
(pattern_unit << rec_2).move((0, -3.5/factor))
(pattern_unit << rec_2).move((-5/factor, 2.5/factor))
pattern_array = gf.components.array(pattern_unit, columns=15*factor, rows=15*factor, column_pitch=12/factor, row_pitch=12/factor)
# pattern_array.show()
pattern_temp = gf.Component()
(pattern_temp << pattern_array).rotate(45)
```

```python
# boolean operation
pattern = gf.Component()
temp = gf.boolean(A = frame, B = pattern_temp,  operation="not", layer1=(1, 0), layer2=(2, 0), layer=(1, 0))
pattern << temp
pattern.show()
```
