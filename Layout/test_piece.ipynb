```python
%reset -f
# test piece for non-hookean metamaterial
import gdsfactory as gf
import import_ipynb
from datetime import datetime
import sys
import math
sys.path.append("/Users/bubble/Desktop/Project/Buckling/non_hookean/Layout")
from gdsfactory.generic_tech import get_generic_pdk
PDK = get_generic_pdk()
PDK.activate()
```

```python
# get the horizontal distance caused by KOH etching.
def KOHSize(height):
    size = height / math.tan(math.radians(54.74))
    return round(size, 2)

# print(KOHSize(525)*2)
```

```python
die = gf.Component()
piece = gf.components.rectangle(size=(20000, 20000), layer=(21, 0))
cell = gf.components.rectangle(size=(15000, 15000), layer=(20, 0))
(die << piece).move((-10000, -10000))
(die << cell).move((-7500, -7500))
widths = [750, 500, 250]
gap = (KOHSize(525)*2+widths[0])/2 + 950/2
for i in range(len(widths)):
    width = widths[i]
    length = 12000
    koh_width = KOHSize(525)*2+width
    koh_length = length + KOHSize(525)*2
    rect = gf.components.rectangle(size=(length, width), layer=(1, 0))
    koh_rect = gf.components.rectangle(size=(koh_length, koh_width), layer=(3, 0))
    if i > 0:
        gap = gap + koh_width+950
    (die << rect).move((-length/2, -width/2+gap))
    (die << koh_rect).move((-koh_length/2, -koh_width/2+gap))

widths = [1000, 2000]
gap = (KOHSize(525)*2+widths[0])/2 + 950/2
for i in range(len(widths)):
    width = widths[i]
    length = 12000
    koh_width = KOHSize(525)*2+width
    koh_length = length + KOHSize(525)*2
    rect = gf.components.rectangle(size=(length, width), layer=(1, 0))
    koh_rect = gf.components.rectangle(size=(koh_length, koh_width), layer=(3, 0))
    if i > 0:
        gap = gap + koh_width + 950
    (die << rect).move((-length/2, -width/2-gap))
    (die << koh_rect).move((-koh_length/2, -koh_width/2-gap))

die.show()
suffix = datetime.now().strftime("%Y%m%d_%H%M")
# die.write_gds(f"wafer_KOH_Test{suffix}.gds", with_metadata=False)

```
