```python
# get the horizontal distance caused by KOH etching.
def KOHSize(height):
    size = height / math.tan(math.radians(54.74))
    return round(size, 2)

# print(KOHSize(525)*2)
```

```python
die = gf.Component()
piece = gf.components.rectangle(size=(20000, 20000), layer=(21, 0))
cell = gf.components.rectangle(size=(15000, 15000), layer=(20, 0))
(die << piece).move((-10000, -10000))
(die << cell).move((-7500, -7500))
width = 5000
length = 10000
rec = gf.components.rectangle(size=(length, width), layer=(1, 0))
(die << rec).move((-length/2, -width/2 + 5000))
(die << rec).move((-length/2, -width/2 - 5000))


die.show()
suffix = datetime.now().strftime("%Y%m%d_%H%M")
die.write_gds(f"wafer_SiN_Test{suffix}.gds", with_metadata=False)

```
